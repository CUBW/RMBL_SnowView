```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
# open a few tiles from test
file_path = "../data/512_splits_4_channel/test_0_images.npz"

dataset = np.load(file_path)

img_arrays = dataset.files

# open first image
img = dataset[img_arrays[20]]

# show image
plt.imshow(np.transpose(img[:3, :, :], (1,2,0)))
```

```python
# load in model
with open("3_feature_model.pkl", 'rb') as f:
    rf = pickle.load(f)
```

```python
#flatten image
img_3 = img[:3, :, :]

img_trans = np.transpose(img_3, (1,2,0))
img_flat = np.reshape(img_trans, (-1,3))


# run predictions
pred_flat = rf.predict(img_flat)

pred_reshape = np.reshape(pred_flat, (512,512))
plt.imshow(pred_reshape, cmap='gray')
```

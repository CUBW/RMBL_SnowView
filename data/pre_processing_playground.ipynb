```python
import cv2
import numpy as np
import rasterio
import os
import csv
import matplotlib.pyplot as plt
import pandas as pd

from rasterio.plot import show
```

```python
#generate an nparray of all of the shapes of the images
shapes = pd.DataFrame(columns=['Shape'])

#open the csv
csv_file = open('./2019-20/2019-20.csv')
csv_reader = csv.reader(csv_file, delimiter=',')
csv_reader.__next__()#skip the header
#loop through the csv
for row in csv_reader:
    # print(row[2])
    #open the image
    img = rasterio.open(row[2])
    #get the shape of the image
    shapes = shapes._append({'Shape': img.shape}, ignore_index=True)
    
```

```python
import matplotlib.pyplot as plt

# Extract the shape values from the DataFrame
shape_values = shapes['Shape']

# Create a histogram
plt.hist(shape_values, bins=10)

# Add labels and title
plt.xlabel('Shape')
plt.ylabel('Frequency')
plt.title('Histogram of Shapes')

# Display the histogram
plt.show()
```

```python
shapes['Shape'].describe()
shapes['Shape'].mode()
shape_freq = shapes['Shape'].value_counts()
print(shape_freq)
```
